```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as dt
import sklearn.cluster as cl
```

```python
IRIS = dt.load_iris()

X = IRIS.data
Y = IRIS.target
```

```python
dbscan = cl.DBSCAN(eps = 0.41, min_samples = 2)
dbscan.fit(X)
Labels = dbscan.labels_
```

```python
Noises = Labels==-1
```

```python
plt.scatter(X[:, 2], X[:, 3], s = 18, c = Labels)
plt.scatter(X[Noises, 2], X[Noises, 3], s = 18, c = 'r', marker = 'x')
plt.xlabel('X3')
plt.ylabel('X4')
plt.show()
```
